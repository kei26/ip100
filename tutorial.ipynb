```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imori.png')

plt.imshow(img)
plt.show()
```

```python
img = img[...,::-1]
plt.imshow(img)
plt.show()
```

```python
print('min = ', img.min())
print('max = ', img.max())
```

```python
r = img[..., 0]
g = img[..., 1]
b = img[..., 2]
plt.subplot(1, 3, 1)
plt.title('R')
plt.imshow(r, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('G')
plt.imshow(r, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('B')
plt.imshow(r, cmap='gray')
plt.show()
```
